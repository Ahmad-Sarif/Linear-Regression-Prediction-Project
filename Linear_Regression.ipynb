# Analisis Regresi dan Korelasi antara jumlah penduduk dengan jumlah energi yang terjual di wilayah Jawa Tengah pada tahun 2015-2019 dan melakukan prediksi terhadap energi yang terjual pada tahun 2020-2022.

1) Melakukan perhitungan regresi linier dan korelasi linier sederhana

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Data (X sebagai variabel independen, Y sebagai variabel dependen)
X = np.array([33774141, 34019095, 34257865, 34490835, 34718204]) # X adalah Data Penduduk Jawa Tengah 2015-2019
Y = np.array([20408190000,  21674840000, 21057040000, 23558010000, 24750620000]) # Y adalah Energi Terjual di Jawa Tengah 2015-2019

# Menghitung koefisien kemiringan (slope) dan intersep (intercept)
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

# Mencetak hasil regresi
print(f"Kemiringan (slope): {slope:.2f}")
print(f"Intersep (intercept): {intercept:.2f}")
print(f"Koefisien determinasi (R-squared): {r_value**2:.2f}")
print(f"Kesalahan standar estimasi: {std_err:.2f}")
print(f"Nilai p: {p_value:.4f}")

if p_value <= 0.05:
  print("Hipotesis nol ditolak")
else:
  print("Hipotesis nol diterima")
```

2) Menentukan Prediksi dan juga esitmasi interval prediksi tahun 2020

```python
import statsmodels.api as sm

# Membuat model regresi menggunakan statsmodels
X_Model = sm.add_constant(X)  # Menambahkan konstanta ke model
model = sm.OLS(Y, X_Model).fit()

# Mencetak estimasi interval untuk prediksi pada tahun 2020
new_X1 = 36516035  # Jumlah Penduduk untuk tahun 2020
new_X1_with_const = np.array([1, new_X1])  # Menambahkan konstanta
predicted_Y1 = model.predict(new_X1_with_const)[0]  # Menghitung prediksi

# Mencetak estimasi interval
confidence_interval = model.get_prediction(new_X1_with_const).conf_int(alpha=0.05)
lower_bound = confidence_interval[0][0]
upper_bound = confidence_interval[0][1]

print(f"Prediksi untuk X = {new_X1}: {predicted_Y1:.2f}")
print(f"Interval kepercayaan (95%): ({lower_bound:.2f}, {upper_bound:.2f})")

```

3) Membuat Diagram pencar terhadap prediksi tahun 2020

```python
# Membuat Diagram Pencar
plt.scatter(X, Y)
plt.xlabel("Jumlah Penduduk ")
plt.ylabel("Jumlah Energi Terjual ")
plt.title("Gambar Diagram Pencar")

# Menampilkan grafik hasil prediksi
plt.plot(new_X1, predicted_Y1, marker='o', linestyle='-', color='red', label='2020')
plt.legend()
plt.show()
```
